# Colors
## how to create colors dictionary

```python
# use one of colormap atribute - hsv

name = pd.DataFrame(dict(zip(['name', 'age'],
                             [['Bohdan', 'Oksana', 'Tamila'], 
                              [40, 40, 13]])))                               # df creation

name_color = dict(zip(name['name'], 
                      plt.get_cmap('hsv')(np.linspace(0.2, 0.5, 3))))        # dict with colors creation

name['color'] = name['name'].apply(lambda x: name_color[x])                  # add colors columns related with values
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

x = np.arange(3)
y = name['age']
color = name['color']

bar = ax.bar(x, y, color=color, width=0.5)

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(name['name'])

ax.set_yticks(list(range(0, 60, 10)))
ax.set_yticklabels(list(range(0, 60, 10)))

# create legend labels

leg_labels = dict(zip(x, bar))

# add bars labels

ax.bar_label(bar, labels=y, label_type='center', fontsize=12)

ax.legend(leg_labels.values(), leg_labels.keys())

plt.show()
```

# Mpl style

```python
styles = plt.style.available
```

```python
fig, axs = plt.subplots(figsize=(25, 40), ncols=3, nrows=9)

x = np.arange(5)
iter_count = -1

for row in range(9):
    for col in range(3):
        iter_count += 1
        if iter_count <=25:
            y = np.random.beta(0.2, 0.7, 5)
            ax = axs[row][col]
            plt.style.use(styles[iter_count])
            ax.plot(x, y)
            ax.set_title(styles[iter_count], fontsize=20)
```
