## work with symmetric distribution

```python
# create two datasets distributed by normal low and Gamma low
ns = 10000             # clients amount
n_mean = 500           # mean income
n_sigma = 1000         # standart deviation
```

```python
# create np array v:
v = np.random.normal(n_mean, n_sigma, ns)
v
```

```python
# create df by v:
v = pd.DataFrame(v, columns=['volume'])
v.head()
```

```python
print(f"mean - {v['volume'].mean()}\n\n"                
      f"median - {v['volume'].median()}\n\n"
      f"moda - {v['volume'].value_counts().nlargest(10)}\n\n"
      f"std_val - {v.volume.std()}\n"
      f"mediana quantile - {np.percentile(v.volume, 50)}\n\n"
      f"0.75 quantile - {np.percentile(v.volume, 75)}\n"
     )
```

```python
v.volume.describe() # use describe method
```

```python
# use seaborn
sns.displot(v)
plt.title('Income distribution by group 1 users')
```

```python
# use pandas
v.volume.hist(bins=100)
plt.title('Income distribution by group 1 users')
```

```python
# kurtosis
kurtosis(v.volume)
```

```python
# asymmetry (skew)
skew(v.volume)
```

```python
# join all metrics in one function for suitable calculations
def my_basic_research(df=v, column='volume'):
    print('Basic metrics')
    print(df[column].describe())
    print("-----------------------------------")
    
    print('Most popular metric values, TOP 5')
    print(df[column].value_counts().nlargest(5))
    print("-----------------------------------")
    
    print('Kurtosis', kurtosis(df[column]))
    print('Skew', skew(df[column]))
    
    sns.displot(df[column])
    plt.title('Income distribution by users')
```

```python
my_basic_research()                 # run function
```

## work with asymmetric distribution

```python
# use function for Gamma parameters distribution search
def gamma_params(mean, std):
    shape = round((mean/std)**2, 4)
    scale = round((std**2)/mean, 4)
    return (shape, scale)
```

```python
shape, scale = gamma_params(n_mean, n_sigma)
df = np.random.gamma(shape, scale, ns)
df = pd.DataFrame(df, columns=['volume'])

# round to tenth
df.volume = round(df.volume, 0)
```

```python
df.head()
```

```python
# use my_research_function
my_basic_research(df = df, column='volume')
```

```python
# unprofitable customers share

print(f"{v[v['volume'] < 0].count()/len(v)}\n"
      f"{df[df['volume'] < 0].count()/len(df)}"
     )
```

```python
# compare general profit by clients with profit more than median in mln
v[v['volume'] >= np.percentile(v.volume, 50)].volume.sum()/10**6
```

```python
df[df['volume'] >= np.percentile(df.volume, 50)].volume.sum()/10**6
```

```python
# get quotes for searching period
start_date = datetime(2018, 1, 1)
end_date = datetime(2019, 8, 31)
```

```python
df = yf.download("GOOG", start=start_date, end=end_date)
```

```python
df.info()
```

```python
df.columns = map(str.lower, df.columns)
df.head()
```

```python
# prepare new metric
df["diff_price"] = df.close - df.open
```

```python
df.diff_price.plot()
plt.title("График разницы между ценой закрытия и открытия")
```

```python
def my_basic_research(df=df, column = "diff_price"):
    print("Базовые метрики")
    print(df[column].describe())
    print("------------------------------------")
    
    print("Самые популярные значения метрики, топ 5")
    print(df[column].value_counts().nlargest(5))
    print("------------------------------------")
    
    print("Эксцесс ", kurtosis(df[column]))
    print("Ассиметрия ", skew(df[column]))
    
    sns.distplot(df[column])
    plt.title("Распределение метрики")
```

```python
my_basic_research(column = "diff_price")
```

```python
sm.qqplot(df.diff_price, fit = True)
plt.title("QQ plot")
plt.show()
```

```python
def my_norm_confidence(df=df, column = "diff_price", alpha = 0.90):
    interval = stats.norm.interval(alpha, loc=df[column].mean(), scale=df[column].std())
    return interval
```

```python
confidence = my_norm_confidence()
print(confidence)
```

```python
df.diff_price.plot()
plt.axhline(y=confidence[1], color="g", linestyle="-")
plt.axhline(y=confidence[0], color="r", linestyle="-")
plt.title("Доверительный интервал")
```

```python
# add column where diff price values are big than upper confidence
df["anomaly_big"] = np.where(df.diff_price > confidence[1], 1,0) 
```

```python
df[df.anomaly_big == 1].describe()
```

```python
from sklearn.datasets import load_diabetes
```

```python
def find_anomaly(df, column):
    upper_level = df[column].mean() + 3*df[column].std()
    level_90 = np.percentile(df[column], 90)
    anomaly = df[(df[column] >= upper_level)^(df[column] >= level_90)]
    return anomaly,upper_level,level_90
```

```python
# load dataset
medical = load_diabetes()
```

```python
# load features fro analysis
df = pd.DataFrame(medical.data,columns = medical.feature_names)

# add target variable - real estate prices
df["health_status"] = medical.target

# fit columns to lower register
df.columns = map(str.lower, df.columns)
df.head()
```

```python
anomaly = find_anomaly(df=df, column = "health_status")
len(anomaly[0])/len(df)
```
